# 2. Data Preprocessing:

# Checking for the missing values:

```python
Processed_Flipdata.isnull().sum()
```

# Check Duplicate values:

```python
Processed_Flipdata.duplicated().sum()
```

# Drop Unnecessary Columns:

```python
Processed_Flipdata=Processed_Flipdata.drop(['Unnamed: 0'], axis=1)
```

```python
Processed_Flipdata.head()
```

# Convert object column itno int64:

```python
Processed_Flipdata['Front Camera'] = Processed_Flipdata['Front Camera'].apply(lambda x: int(x[:-2]))
```

```python
Processed_Flipdata['Rear Camera'] = Processed_Flipdata['Rear Camera'].apply(lambda x: int(x[:-2]))
```

# Introducing missing values:

```python
Processed_Flipdata.loc[Processed_Flipdata.sample(frac=0.1).index, ['Memory','RAM','Battery_','Rear Camera','Front Camera','AI Lens','Mobile Height','Prize']] = np.nan
```

```python
Processed_Flipdata.isnull().sum()
```

# Check-Outliers:

# Box_plot for target prize column:

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=Processed_Flipdata, y='Prize')
plt.ylabel('Prize')
plt.title('Box Plot of Prize')
plt.show()
```

# Kernel_Density Plot for target prize column:

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=Processed_Flipdata['Prize'], shade=True)
plt.xlabel('Prize')
plt.ylabel('Density')
plt.title('Kernel Density Estimation of Prize')
plt.show()
```

# Remove Outliers:

```python
quantile_low = Processed_Flipdata['Prize'].quantile(0.01)
quantile_high = Processed_Flipdata['Prize'].quantile(0.99)
Processed_Flipdata = Processed_Flipdata[(Processed_Flipdata['Prize']>quantile_low) & (Processed_Flipdata['Prize']<quantile_high)]
```

```python
Processed_Flipdata.head()
```

# One-hot_Encoding:

```python
hot_Processed_Flipdata=pd.get_dummies(Processed_Flipdata, columns= ['Model','Colour','Processor_'],
                                           drop_first=True,dtype=int)
```

# Apply PCA:

```python
from sklearn.decomposition import PCA
```

```python
n_components = 5
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

# Explained Variance Ratio:

```python
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)
```

# Selecting Number of Components:

```python
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
plt.plot(range(1, len(cumulative_variance_ratio)+1), cumulative_variance_ratio, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('PCA: Cumulative Explained Variance Ratio')
plt.show()
```

# Statistical Methods:

# ANOVA (Analysis of Variance):

```python
from scipy.stats import f_oneway
```

```python
groups = [hot_Processed_Flipdata[hot_Processed_Flipdata['Front Camera'] == value]['Prize'] for value in hot_Processed_Flipdata['Front Camera'].unique()]

# Perform ANOVA
f_statistic, p_value = f_oneway(*groups)
print("F-Statistic:", f_statistic)
print("P-Value:", p_value)
```

# correlation between feature columns and target column:

```python
cols = ['Rear Camera','Front Camera','Battery_','RAM','Memory']

plt.figure(figsize=(24, 12))
for index, column in enumerate(cols):
    plt.subplot(3, 3, index + 1)
    sns.regplot(x=hot_Processed_Flipdata[column], y=hot_Processed_Flipdata["Prize"],
                color="black", scatter_kws={'s': 70, 'alpha': 0.5}, line_kws={'color': 'red', 'lw': 3})
    corr = round(hot_Processed_Flipdata[[column, "Prize"]].corr()["Prize"][0], 2)
    plt.title(f"Correlation value is {corr}", pad=10, size=20, fontweight="black")
plt.tight_layout()
plt.show()
```

# Split the Dataset:

```python
from sklearn.model_selection import train_test_split
```

```python
X = hot_Processed_Flipdata[['Rear Camera','Battery_','Front Camera','RAM','Memory']]
Y = hot_Processed_Flipdata[['Prize']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```
